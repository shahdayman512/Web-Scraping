```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
page=requests.get('https://books.toscrape.com/catalogue/category/books/travel_2/index.html')
```

```python
src=page.content
```

```python
soup=BeautifulSoup(src,'lxml')
```

```python
books = soup.find_all('article',class_='product_pod')
```

```python
num_books=len(books)
num_books
```

```python
name_list=[]
price_list=[]
rating_list=[]
availability_list=[]
```

```python
for i in range(num_books):
    book_name=books[i].find('h3').text.strip()
    book_price=books[i].find('p',class_='price_color').text.strip()
    book_rating=books[i].find('p',class_='star-rating')['class'][1]
    book_availability=books[i].find('p',class_='instock availability').text.strip()
    name_list.append(book_name)
    price_list.append(book_price)
    rating_list.append(book_rating)
    availability_list.append(book_availability)
```

```python
book_store_file=pd.DataFrame({'Name':name_list,'Price':price_list,'Rating':rating_list,'Availability':availability_list})
book_store_file.to_csv('book_store.csv',index=True)
```
